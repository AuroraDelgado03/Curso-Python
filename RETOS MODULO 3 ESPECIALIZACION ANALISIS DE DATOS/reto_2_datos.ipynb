```python
# --- RETO 2 ---

# --- Importar ---
import pandas as pd
import numpy as np
print("Pandas y Numpy importados.")

# --- Crear una serie
calificaciones = {'Ana': 10, 'Luis': 7.5, 'Marta': 9.0, 'Juan': 5.5}
s_calificaciones = pd.Series(calificaciones)
print(f"\nSerie original:\n{s_calificaciones}")

# --- Crear una nueva serie sumando 0.5 a la serie original
s_calificaciones_final = pd.Series(s_calificaciones + 0.5)
print(f"\nSerie modificada:\n{s_calificaciones_final}")
```

```python
# --- Crear un DataFrame
data = {
    'producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor'],
    'stock': [20, 150, 80, 35],
    'precio_usd': [1200, 25, 60, 300]
}
df_inventario = pd.DataFrame(data)
print(f"DataFrame original:\n{df_inventario}")

# --- Modificar para que la columna 'producto' sea el nuevo índice de filas
df_inventario.set_index('producto', inplace=True)
print(f"\nDataFrame modificado:\n{df_inventario}")
```
